```python
import requests
import pandas as pd
import json
from pandas import json_normalize 
```

```python
resp = requests.get("https://data.iledefrance.fr/api/records/1.0/search/?dataset=base-comparateur-de-territoires&q=&lang=en&rows=100&facet=dep&refine.dep=75")
print(resp.status_code)
```

```python
json = resp.json()
```

```python
json.keys()
```

```python
type(json['records'][0])
```

```python
x=pd.json_normalize(json['records'])
```

```python
df = pd.DataFrame(x)
```

```python
df=df.drop(columns =['datasetid', 'recordid'])
```
