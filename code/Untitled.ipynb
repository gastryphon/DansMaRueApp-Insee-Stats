```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
import seaborn as sns
```

```python
df0 = pd.read_csv(r'/Users/davidpitoun/RNCP_project/db csv/locationsdf.csv')
```

```python
df10 = pd.read_csv(r'/Users/davidpitoun/RNCP_project/db csv/socioecodf.csv')
```
